#are there patterns in the passenger count across diffrent times of day

```python

data2["pickup_hour"]=pd.to_datetime(data2["tpep_pickup_datetime"]).dt.hour
passenger_count_by_hour=data2.groupby("pickup_hour")["passenger_count"].mean()
pd.DataFrame(data2)
passenger_count_by_hour=passenger_count_by_hour.reset_index()
plt.close("all")
plt.figure(figsize=(12,6))
plt.plot(passenger_count_by_hour["pickup_hour"],passenger_count_by_hour["pickup_hour"],marker="o",linestyle="-",color="b")
plt.title("avarage passenger count by hour of the day")
plt.xlabel("hour of the day")
plt.ylabel("passenger count")
plt.xticks(range(0,24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

#are there any zero distance trips

```python

zero_diatance_trips=(data2["trip_distance"]<=0)
print(f"number of zero trip distance:{len(zero_diatance_trips)}")
print(zero_diatance_trips)
```

#how many trips are completed in a month

```python

data2["pickup_month"]=pd.to_datetime(data2["tpep_pickup_datetime"]).dt.to_period("M")
trips_per_month=data2.groupby("pickup_month").size()
print("number of trips completed in a month")
print(trips_per_month)
```
